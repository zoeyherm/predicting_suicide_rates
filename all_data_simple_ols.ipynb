```python
df = df4.copy()
```

```python
df = df.rename(columns = {'Total\nofficers': 'Total_officers'})
```

```python
df['Total_officers'] = df['Total_officers'].str.replace(',', '')
```

```python
df['Total_officers'] = pd.to_numeric(df['Total_officers'])
```

```python
df['cost'] = df['cost'].str.replace(',', '')
```

```python
df['cost'] = df['cost'].str.replace('$', '')
```

```python
df['cost'] = pd.to_numeric(df['cost'])
```

```python
df = df.rename(columns = 
               {'Total\nofficers': 'Total_officers', 
                'Area in square miles - Land area': "Land area",
                '% With Access eng 1': '% With Access' })


df['officers_per_area'] = df['Total_officers'] / df['Land area']
df['pcp_per_area'] = df['PCP Ratio'] / df['Land area']
df['mhp_per_area'] = df['MHP Ratio'] / df['Land area']
df['income_ratio'] = df['cost'] / df['med_house_income']
```

```python
df['care_profs_per_area'] = df['pcp_per_area'] + df['mhp_per_area']
```

```python
df['Care Prof Ratio'] = df['PCP Ratio'] + df['MHP Ratio']
```

```python
df['log_rate'] = np.log(df['rate'])
```
